```python
filepath = "receipts.txt"
with open(filepath) as f:
    receipts_str = f.read()
print(receipts_str[:500])
```

```python
date = r"\d{2}-\d{2}-\d{4}"
text_to_search = receipts_str
DATE = re.findall(date, text_to_search)
DATE
```

```python
quantity = r"Quantity +(\d+)"
text_to_search = receipts_str
QUANTITY = re.findall(quantity, text_to_search)
QUANTITY
```

```python
amount = r"Total Amount +(\d+.\d{2}) â‚¬"
text_to_search = receipts_str
AMOUNT = re.findall(amount, text_to_search)
AMOUNT
```

```python
receipts_list = receipts_str.split("====================================")
len(receipts_list)
```

```python
receipts_dict = {
    "date": [],
    "quantity": [],
    "total_amount": []
}
```

```python
receipts_dict = {
    "date": DATE,
    "quantity": QUANTITY,
    "total_amount": AMOUNT
}
```

```python
receipts_dict_df = pd.DataFrame(receipts_dict)
receipts_dict_df
```

```python
for receipt in receipts_list:
    print(receipt)
```

```python
print(f"date: {DATE}, total_amount: {AMOUNT}, quantity: {QUANTITY}")
```

```python
receipts_dict_df["date"] = pd.to_datetime(receipts_dict_df["date"], format="%d-%m-%Y")
receipts_dict_df.head(3)
```

```python
receipts_dict_df.sort_values('date', inplace=True)
receipts_dict_df.head(3)
```

```python
receipts_dict_df = receipts_dict_df.set_index('date')
```

```python
receipts_dict_df["quantity"] = pd.to_numeric(receipts_dict_df["quantity"])
receipts_dict_df["total_amount"] = pd.to_numeric(receipts_dict_df["total_amount"])
```

```python
receipts_dict_df['total_amount'].plot(figsize=(14,5))
receipts_dict_df.plot(figsize=(14,5))
```
